```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as gambar
```

```python
def convolution2D(image, kernel):
  # Mengimport scipy dikarenakan meningkatkan performa dalam konvolusi 
  # yang menggunakan algoritma yang lebih efisien
  from scipy.signal import convolve2d
  
  # Mode same yang tertulis dibawah ini digunakan untuk memastikan bahwa
  # ukuran output nya sesuai/sama dengan ukuran image yang diinput
  hasil_konvolusi = convolve2d(image, kernel, mode='same')
    
  return hasil_konvolusi
```

```python
img_read = cv.imread('/content/drive/MyDrive/Andre/female.tiff')
img_gray = cv.cvtColor(img_read, cv.COLOR_BGR2GRAY)
```

```python
# inisialisasi kernel
kernel_sharpen = np.array([ [0, -1, 0], 
                   [-1, 5, -1], 
                   [0, -1, 0] ])

# melakukan konvolusi
output = convolution2D(img_gray, kernel_sharpen)

# menampilkan gambar
cv2_imshow(img_gray)
print('Hasil')
print(img_gray)
```

```python
kernel = np.array([ [0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0] ])

# Mengaplikasikan filter ke dalam image menggunakan cv dan kernel
image_sharpen = cv.filter2D(img_gray, -1, kernel)

plt.imshow(cv.cvtColor(image_sharpen, cv.COLOR_GRAY2RGB))
```
