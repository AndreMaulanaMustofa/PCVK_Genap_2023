```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Andre/morphology.png')

height = img.shape[0]
width = img.shape[1]

def dilasi(img):
    img_dilasi = np.maximum.reduce([
        img[:-2, :-2], img[:-2, 1:-1], img[:-2, 2:],
        img[1:-1, :-2], img[1:-1, 1:-1], img[1:-1, 2:],
        img[2:, :-2], img[2:, 1:-1], img[2:, 2:]
    ])
    return img_dilasi

hasil = dilasi(img)
plt.subplot(121),plt.imshow(img), plt.title('Citra Awal')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(hasil), plt.title('Citra Hasil')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('/content/drive/MyDrive/Andre/morphology.png')

sed = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))

img_dilasi = cv2.dilate(img, sed, iterations=1)

print("Structuring Element Dilasi (SED):")
print(sed)
plt.subplot(121),plt.imshow(img), plt.title('Citra Awal')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_dilasi), plt.title('Citra Hasil')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_j = cv2.imread('/content/drive/MyDrive/Andre/j.png')

def erosi(img):
    # Membuat Structuring Element square 5 x 5
    se = np.ones((5, 5), np.uint8)

    # Operasi erosi
    eroded = cv2.erode(img, se)

    return eroded

eroded = erosi(img_j)

plt.subplot(121),plt.imshow(img_j), plt.title('Citra Awal')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(eroded), plt.title('Citra Hasil')
plt.xticks([]), plt.yticks([])
plt.show()
```
