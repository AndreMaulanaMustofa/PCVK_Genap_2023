```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gamma Correction
print ('Gamma Correction pada Citra ')
print ('---------------------------------')
try:
  gamma =  int(input('Masukkan Nilai Gamma: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/Andre/female.tiff')
gamma_correct = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      gamma_correct[y,x,c] = np.clip(255*pow(original[y,x,c]/255,(1/gamma)),0,255)

final_frame = cv.hconcat((original, gamma_correct))
cv2_imshow(final_frame)
```

```python
# Depth Image
print ('Depth Image pada Citra ')
print ('---------------------------------')
try:
  depth =  int(input('Masukkan Nilai depth: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/Andre/female.tiff', cv.IMREAD_GRAYSCALE)
depth_correct = np.zeros(original.shape, original.dtype)
level = 255 / (pow(2,depth)-1)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    depth_correct[y,x] = np.clip(round(original[y,x] / level) * level ,0,200)

final_frame = cv.hconcat((original, depth_correct))
cv2_imshow(final_frame)
```

```python
import glob
from math import log10, sqrt

original = cv.imread('/content/drive/MyDrive/Andre/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/Andre/noises/*'):
  n = cv.imread(img)
  cv_img.append(n)

dst = cv_img[0]
noise_number = 4 # jumlah citra array yang di average
for i in range(noise_number):
  if i == 0 :
    pass
  else:
    alpha = 1.0 / (i + 1)
    beta = 1.0 - alpha
    dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(original, compressed):
  mse = np.mean((original - compressed) ** 2)
  if(mse == 0):
    return 100
  max_pixel = 255.0
  psnr = 20* log10(max_pixel / sqrt(mse))
  return psnr

compressed = dst
psnr = PSNR(original, compressed)

print('Nilai PSNR adalah ', psnr, '(dB)')
galaxy_frame = cv.hconcat((original, dst))
cv2_imshow(galaxy_frame)
```

```python
# Jumlah Citra 30

import math
import cv2
import numpy as np

original = cv2.imread('/content/drive/MyDrive/Andre/noises/1.jpg')
contrast = cv2.imread('/content/drive/MyDrive/Andre/noises/29.jpg', 1)

def psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return 100
    PIXEL_MAX = 255.0
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))

d = psnr(original, contrast)
print('Psnr value',d,'(dB)')

final_frame = cv.hconcat((original, contrast))
cv2_imshow(final_frame)
```
